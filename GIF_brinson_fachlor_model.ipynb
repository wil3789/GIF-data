```python
'''
    Now we need to make our sector performance comparable to the ACWI ETF performance, 
    so we regroup GICS_sector into 6 new sectors which is the same as GIF_sector and 
    recalculate it's weighting and performance.
    
    Sector Abbreviation:
    Energy, Industrial, Material, &utility = eimu
    Consumer Staple & Discretionary = csd
    Financial & Real Estate = fr
    Healthcare = health
    Technology = tech
    Communication Service = cs
'''
```

```python
GIF_eimu_weight = 4.254405 + 8.035865 + 4.646547 + 5.194960
GIF_csd_weight = 6.010491 + 9.332454
GIF_fr_weight = 16.282619 + 3.428905
GIF_health_weight = 16.134985
GIF_tech_weight = 17.270550
GIF_cs_weight = 4.524107
```

```python
ACWI_eimu_weight = 5.21 + 9.68 + 4.51 + 3.14
ACWI_csd_weight = 11.16 + 7.65
ACWI_fr_weight = 15.4 + 2.82
ACWI_health_weight = 12.34
ACWI_tech_weight = 20.43
ACWI_cs_weight = 7.64

ACWI_eimu_return = (5.21/ACWI_eimu_weight)*49.80 \
                    +(9.68/ACWI_eimu_weight)*(-12) \
                    +(4.51/ACWI_eimu_weight)*(-13.8) \
                    +(3.14/ACWI_eimu_weight)*(5.3)
ACWI_csd_return = (11.16/ACWI_csd_weight)*(-23.5) \
                    +(7.65/ACWI_csd_weight)*(-4.1)
ACWI_fr_return = (15.40/ACWI_fr_weight)*(-14.3) \
                    +(2.82/ACWI_fr_weight)*(-13.63)
ACWI_health_return = -12.1
ACWI_tech_return = -22.8
ACWI_cs_return = -30.4
```

```python
'''Make array to make easier to calculate'''
l_GIF_weight = np.array([GIF_eimu_weight,GIF_csd_weight,GIF_fr_weight,GIF_health_weight,GIF_tech_weight,GIF_cs_weight])
l_GIF_return = np.array([ener_ind_mat_uti_return,consumer_return,fin_estate_return,healthcare_return,technology_return,comm_serv_return])
l_ACWI_weight = np.array([ACWI_eimu_weight,ACWI_csd_weight,ACWI_fr_weight,ACWI_health_weight,ACWI_tech_weight,ACWI_cs_weight])
#use Vanguard return as a temporary solution
l_ACWI_return = np.array([ACWI_eimu_return,ACWI_csd_return,ACWI_fr_return,ACWI_health_return,ACWI_tech_return,ACWI_cs_return])
benchmark_return = -17.32
```

```python
def brinson_fachler(wi, Wi, Ri, Bi, B):
    A = 0  #allocation effect
    wi = wi/100
    Wi = Wi/100
    Ri = Ri/100
    Bi = Bi/100
    B = B/100
    for k in range(len(wi)):
        Ai = (wi[k] - Wi[k])*(Bi[k] - B)
        A += Ai
    
    S = 0  #selection effect
    for k in range(len(wi)):
        Si = Wi[k]*(Ri[k] - Bi[k])
        S += Si
    
    I = 0  #interaction effect
    for k in range(len(wi)):
        Ii = (wi[k] - Wi[k])*(Ri[k] - Bi[k])
        I += Si
    
    return A*100, S*100, I*100
```

```python
A, S, I = brinson_fachler(l_GIF_weight, l_ACWI_weight, l_GIF_return, l_ACWI_return, benchmark_return)
print(A, S, I)
```
